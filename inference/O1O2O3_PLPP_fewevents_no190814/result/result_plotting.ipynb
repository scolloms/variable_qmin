```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import corner as corner
```

```python
result=h5py.File('bbh_mass_two_component_primary_mass_ratio_redshift_powerlaw_result.hdf5')
```

```python
result['posterior'].keys()
```

```python
important_result = [result['posterior']['alpha'], result['posterior']['beta'], result['posterior']['mpp'],\
                    result['posterior']['mmin'], result['posterior']['mmax'], result['posterior']['sigpp']]
```

```python
corner_kwargs = dict(
    bins=32,
    smooth=0.9,
    color="darkslateblue",
    #quantiles=[0.16, 0.84],
    levels=(1 - np.exp(-0.5), 1 - np.exp(-2), 1 - np.exp(-9 / 2.0)),
    plot_density=True,
    plot_datapoints=True,
    fill_contours=True,
    show_titles=True,
    hist_kwargs=dict(density=True),
    label_kwargs=dict(fontsize=25),
    title_kwargs=dict(fontsize=18),
    labels=[r'$\alpha$',r'$\beta$',r'$\gamma$', r'$m_{\rm min}$', r'$m_{\rm max}$', r'$\zeta$'],
)
```

```python

plt.rcParams.update({'font.size': 15})
corner.corner(np.swapaxes(important_result, 0,1), **corner_kwargs)
```
